```python
# import our text mining toolkit
import text_mining_toolkit as tmt
```

```python
cr = tmt.corpus_reader.CorpusReader(directory_of_files="data_sets/recipes/txt/", text_filename_pattern="??.txt")
#cr = tmt.corpus_reader.CorpusReader(directory_of_files="data_sets/mystery_corpus_01/txt/", text_filename_pattern="??.txt")
#cr = tmt.corpus_reader.CorpusReader(directory_of_files="data_sets/iraq_inquiry/txt/", text_filename_pattern="the-report*.txt")
```

```python
# wordcount to wordcloud
# pipeline

# get all text from corpus
a = cr.get_all_text()

# simplify whitespace (remove newlines)
b = tmt.text_processing.simplify_whitespace(a)

# only keep alphanumeric characters, removes punctuation
c = tmt.text_processing.keep_alphanumeric(b)

# make lowercase
d = tmt.text_processing.to_lowercase(c)

# split into words list
dl = tmt.word_processing.split_text_into_words(d)

# remove stop words
el = tmt.word_processing.remove_stop_words(dl, "./stopwords/minimal-stop.txt")

# only keep words with min length 5
fl = tmt.word_processing.keep_words_min_length(el,4)

# build n-grams
gl = tmt.word_processing.build_ngrams_from_words(fl,2)

#print(b, '\n')
#print(c, '\n')
#print(dl, '\n')
#print(el, '\n')
#print(fl, '\'n)
#print(gl, '\n')
```

```python
# word cloud
wc = tmt.word_processing.count_word_frequency(gl)
tmt.visualisation.plot_wordcloud(wc, most_common=None)
```
